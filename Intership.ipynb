```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# a function called testing that reads the gct files into a pandas dataframe

def testing(file, summary):
    df = pd.read_csv(file, sep = "\t", skiprows = 2)
    if summary == True:
        print("This data has " + str(len(df)) + " rows")  # prints the # of rows
        print("This data has " + str(len(df.columns)) + " columns")  #prints the # of columns
    return df
```

```python
# the fist task of opening the fist gct file

input_testing1 = 'BRCA_minimal_60x19.gct'
input_summary1 = True
a = testing(input_testing1, input_summary1)
print(a)
```

```python
# the second task of oepning the second file

input_testing2 = 'BRCA_large_20783x40.gct'
input_summary2 = False
b = testing(input_testing2, input_summary2)
print(b)
```

```python
# the first graph of A7-A0DB-normal

data1_points = []
for index in range(len(b)):
    data1_points.append(b.loc[index]["A7-A0DB-normal"])  # strong the values into a list which is then transformed it into a dataframe
data1 = pd.DataFrame(data1_points)


Q1_1 = data1.quantile(0.25)
Q3_1 = data1.quantile(0.75)
IQR1 = Q3_1 - Q1_1
data1 = data1[~((data1 < (Q1_1 - 1.5 * IQR1)) | (data1 > (Q3_1 + 1.5 * IQR1))).any(axis=1)]  # elimination of any outliers within the data


fig, ax = plt.subplots() 


arr = ax.hist(data1, ec = 'white')
# print(arr)

ax.set_xlim(0, 5000)

plt.title("A7-A0DB-normal")
plt.show()
```
